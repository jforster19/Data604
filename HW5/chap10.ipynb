Some of the estimates are in a form Pandas doesn't recognize as numbers, but we can coerce them to be numeric.

```python
table1.plot()
decorate(xlim=[-10000, 2000], xlabel='Year', 
         ylabel='World population (millions)',
         title='Prehistoric population estimates')
plt.legend(fontsize='small');
```

We can use `xlim` to zoom in on everything after Year 0.

```python
fake_rng = [0,100,250,500,750,1000,1250,1500,1750,2000]
exp_rng = {x:math.exp(x-0.01*x-1.5)+275 for x in range(0,10)}
#exp_rng = {x:math.exp(x-0.001*x-0.2) for x in range(0,10)}
long_exp_rng = {x:math.exp(x-0.01*x-1)+200 for x in range(0,10)}
test_series = pd.Series(exp_rng,name='Model')
test_series.index = fake_rng
test_series2 = pd.Series(long_exp_rng,name='2^X')
test_series2.index = fake_rng
```

```python
table1.plot()
test_series.plot(style='--',label='Exp1')
test_series2.plot(style='--',label='Exp2')
decorate(xlim=[0, 1960], xlabel='Year', 
         ylabel='World population (millions)',
         title='CE population estimates against Exponential')
plt.legend(fontsize='small');
```

The exponential variant identified here appears to fairly closely resemble the initial population estimates predicted from Year 0 through 1940. It is unlikely this function will properly account for more present day population numbers as it is beginning to approach the assymptote and will not level off.

```python
cur_exp_rng = {x:math.exp(x-0.01*x-1.5)+275 for x in range(10,77)}
#exp_rng = {x:math.exp(x-0.001*x-0.2) for x in range(0,10)}
cur_exp_rng2 = {x:math.exp(x-0.01*x-1)+200 for x in range(10,77)}
cur_exp_series = pd.Series(cur_exp_rng,name='Exp1')
cur_idx = np.linspace(1950,2016,67)
cur_exp_series.index = cur_idx
cur_exp_series2 = pd.Series(cur_exp_rng2,name='Exp2')
cur_exp_series2.index = cur_idx
```

```python
# Solution goes here
table2['census'].plot()
cur_exp_series.plot(style='--',color='orange',label='Exp e^x-0.01x-0.2)+200')
cur_exp_series2.plot(style='--',color='green',label='Exp e^x-0.01x-1)+200')
plt.plot();
```
